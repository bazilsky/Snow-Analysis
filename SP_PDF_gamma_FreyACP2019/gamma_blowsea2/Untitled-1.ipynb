```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# read in .mat file 
import scipy.io as sio

filename = 'U1104_8cm_1min.mat'
filepath = '/Users/ananth/Desktop/bas_scripts/DATA_SETS/mosaic/newdata_with_metcity/'
mat_contents = sio.loadmat(filepath+filename)

```

```python
mat_contents.keys()
```

```python
U10m = mat_contents['U10m_ex']
# remove all nan values
U10m = U10m[~np.isnan(U10m)]
```

```python
np.max(U10m)
```

```python
# plot histogram of velocity data with grid in fivethirty eight style and normalised y axis
plt.figure(figsize=(12, 8))
plt.style.use('fivethirtyeight')
plt.hist(U10m, bins=100, density=True, edgecolor = 'black', linewidth=1.2)
plt.xlabel('U10m (m/s)', fontsize = 18)
plt.ylabel('Probability', fontsize = 18)
#plt.title('Wind Speed Distribution', fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.grid(True)

plt.show()
```
