```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.io.shapereader as shpreader
import cartopy.io.img_tiles as cimgt

import matplotlib as mpl

mpl.rcParams['font.family'] = 'sans-serif'
mpl.rcParams['font.sans-serif'] = ['Times New Roman']
```

```python
import random
# read in latlon mosaic csv file 

mosaic_data = np.loadtxt('mosaic_latlon.csv', delimiter=',')
nice_data = np.loadtxt('NICE2015_latlon.csv', delimiter=',')

mosaic_lat = mosaic_data[:,0]
mosaic_lon = mosaic_data[:,1]

flag1 = (mosaic_lat !=np.nan) & (mosaic_lon !=np.nan)
flag2 = (mosaic_lon !=np.nan)

mosaic_lat = mosaic_lat[flag1]
mosaic_lon = mosaic_lon[flag1]


nice_lat = nice_data[:,0]
nice_lon = nice_data[:,1]


station_lat = [82.3,71.2906,78.9067,81.36]
station_lon = [-62.3,-156.5,11.889, -16.4]




station_name = ['Alert', 'Utqiagvik', 'Zepplin', 'Villum']
#station_name = ['Alert', '', 'Zepplin', 'Villum']
station_symbol = ['^', 'h', 'o', 'd']

random.seed(100)
# choose 8 random colors 
station_color = ['r', 'b', 'g', 'olive']
#station_color = [(r,g,b) for i in range(4) r, g, b = random.randint(0,255), g = random.randint(0,255), b = random.randint(0,255)] 
station_color = [(random.randint(0,255)/255.0, random.randint(0,255)/255.0, random.randint(0,255)/255.0) for i in range(4)]
# data from past references
study_name = ['Ewert et al 2013', 'Krnavek et al 2012', 'Cambridge Bay', 'Eureka', 'Resolute Bay', 'Hudson Bay', 'Cape Bathurst']
#study_name = ['Ewert et al 2013', '', 'Cambridge Bay', 'Eureka', 'Resolute Bay', 'Hudson Bay', 'Cape Bathurst']
study_position = [[71.3,-159.8],[71.4,-156.5],[69.03,-105.12], [80.08,-86.79],[74.7,-95.6],[58.46,-93.5],[69.63,-126.11]]
study_position = [[71.9,-161.8],[72.8,-158],[69.03,-105.12], [80.08,-86.79],[74.7,-95.6],[58.46,-93.5],[69.63,-126.11]]

study_color = [(random.randint(0,255)/255, random.randint(0,255)/255, random.randint(0,255)/255) for i in range(7)]

# just for Nandan et al 2017
#study_position = [[71.3,-156.8],[71.4,-156.5],[69.03,-105.12]]
# pick 19 random colors using the randomized funciton in python for each run 



```

```python
# plot latlon data onto a map of the arctic make it artistic

fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection=ccrs.NorthPolarStereo())
#ax = plt.axes(projection = ccrs.Mollweide())
ax.set_extent([-180, 180, 57, 90], ccrs.PlateCarree())
ax.coastlines(resolution='50m', color='black', linewidth=1)

ax.add_feature(cfeature.OCEAN, facecolor='white', alpha=0.2)
ax.add_feature(cfeature.LAND, facecolor='black', alpha=0.7)
#ax.add_feature(cfeature.LAKES, facecolor='lightblue')
#ax.add_feature(cfeature.RIVERS, facecolor='lightblue')
# plot regions of ice 
#ax.add_feature(cfeature.ICE_SHEETS, facecolor='white')

ax.gridlines()

# plot latlon data onto a map of the arctic make it artistic

ax.plot(mosaic_lon, mosaic_lat, marker = 'h', color='orange', markersize=2, transform=ccrs.PlateCarree(), label = 'MOSAIC')
ax.plot(nice_lon, nice_lat, marker = 'h',  color='maroon', markersize=2, transform=ccrs.PlateCarree(), label = 'N-ICE2015')

# plot station locations
for i in range(len(station_lat)):
    ax.plot(station_lon[i], station_lat[i], marker = 'h', color=station_color[i], markersize=15, transform=ccrs.PlateCarree(), label = station_name[i])

for i in range(len(study_position)):
    ax.plot(study_position[i][1], study_position[i][0], marker = 'h', color=study_color[i], markersize=15, transform=ccrs.PlateCarree(), label = study_name[i])

# Plot station names
for i in range(len(station_lat)):
    if station_name[i] == 'Utqiagvik':
        ax.text(station_lon[i] + 20, station_lat[i] - 4, station_name[i], transform=ccrs.PlateCarree(), color = 'orange',fontsize=16, fontstyle = 'oblique', fontweight = 'bold')
    else:
        ax.text(station_lon[i] + 0.5, station_lat[i] + 0.5, station_name[i], transform=ccrs.PlateCarree(), color = 'orange',fontsize=16, fontstyle = 'oblique', fontweight = 'bold')

# Plot study names
for i in range(len(study_name)):
    if study_name[i] == 'Ewert et al 2013':
        ax.text(study_position[i][1] - 0.5, study_position[i][0] - 1, study_name[i], transform=ccrs.PlateCarree(),color = 'orange', fontsize=16, fontstyle = 'oblique', fontweight = 'bold')
    elif study_name[i] == 'Krnavek et al 2012':
        ax.text(study_position[i][1] - 0.5 , study_position[i][0] + 1, study_name[i], transform=ccrs.PlateCarree(), color = 'orange',fontsize=16, fontstyle = 'oblique', fontweight = 'bold')
    else:
        ax.text(study_position[i][1] + 0.5, study_position[i][0] + 0.5, study_name[i], transform=ccrs.PlateCarree(), color = 'orange',fontsize=16, fontstyle = 'oblique', fontweight = 'bold')


#ax.legend(loc='upper right', fontsize=18)
legend = ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3, fontsize=18)
legend.get_frame().set_edgecolor('black')
```
